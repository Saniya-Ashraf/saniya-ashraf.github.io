```python
import pandas as pd
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
titanic = pd.read_csv('datasets/titanic.csv')
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

## 1. Scatterplot (Numerical - Numerical)

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['smoker'],style=tips['sex'],size=tips['size'])
```

## 2. Bar Plot (Numerical - Categorical)

```python
titanic.head()
```

```python
sns.barplot(x=titanic['Pclass'],y=titanic['Age'],hue=titanic['Sex'])
```

## 3. Box Plot (Numerical - Categorical)

```python
sns.boxplot(x=titanic['Sex'],y=titanic['Age'],hue=titanic['Survived'])
```

## 4. Distplot (Numerical - Categorical)

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
```

```python
sns.displot(data=titanic[titanic['Survived'] == 0], x='Age', hue="Pclass", kind="kde")
sns.displot(data=titanic[titanic['Survived'] == 1], x='Age', kind="kde")

```

```python
sns.kdeplot(data=titanic[titanic['Survived'] == 0], x='Age', hue="Pclass")
sns.kdeplot(data=titanic[titanic['Survived'] == 1], x='Age', hue = "Pclass")
```

## 5. HeatMap (Categorical - Categorical)

```python
titanic.head()
```

```python
sns.heatmap(tips.corr())
```

```python
(titanic.groupby('Pclass').mean()['Survived']*100).plot(kind='bar')
```

## 6. ClusterMap (Categorical - Categorical)

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```

## 7. Pairplot

```python
iris.head()
```

```python
sns.pairplot(iris,hue='petal_length')
```

## 8. Lineplot (Numerical - Numerical)

```python
flights.head()
```

```python
new = flights.groupby('year').sum().reset_index()
```

```python
sns.lineplot(x=new['year'],y=new['passengers'])
```

```python
flights
```

```python
sns.clustermap(flights.pivot_table(values='passengers',index='month',columns='year'))
```
